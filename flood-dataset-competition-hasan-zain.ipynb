```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/kaggle/input/playground-series-s4e5/train.csv")
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
sns.heatmap(data.corr().round(4), annot=True, cmap='YlOrRd')
```

```python
X_train = data.drop('FloodProbability', axis=1)
y_train = data['FloodProbability']
```

```python
import xgboost as xgb

model = xgb.XGBRegressor(n_estimators=100, max_depth=3, learning_rate=0.1)
model.fit(X_train, y_train)
```

```python
X_test = pd.read_csv('/kaggle/input/playground-series-s4e5/test.csv')
```

```python
X_test.head()
```

```python
predictions = model.predict(X_test)
```

```python
y_test = pd.DataFrame({'id': X_test['id'], 'FloodProbability': predictions})
```

```python
y_test
```

```python
y_test.to_csv('/kaggle/working/predictions.csv', index=False)
```
